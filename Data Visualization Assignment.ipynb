```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
import pandas as pd
```

```python
iris=datasets.load_iris()
```

```python
iris.feature_names
```

```python
iris.feature_names
iris=iris.data
```

```python
df=pd.DataFrame(iris[0,:],columns=['sepal_length'])
```

```python
df['sepal_width']=iris[1,:]
df['petal_length']=iris[2,:]
```

```python
x = iris[:,0]
y = iris[:,1]
z =  iris[:,2]
dx = np.ones(iris[:,0].size)
dy = np.ones(iris[:,1].size)
dz = np.ones(iris[:,2].size)

fig = plt.figure(figsize=(10,8))
ax = Axes3D(fig)
bar3d = ax.bar3d(x, y, z, dx, dy, dz, color='g')

ax.set_title("IRIS Dataset Analysis ")
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
ax.set_zlabel('petal length (cm)')

plt.show()
```
